```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
reward = []
change = []


for i in range(25):
    
    all_doors      = [1, 2, 3]                      # All doors
    reward         = np.random.choice(all_doors)    # Define where the reward sits
    primary_choice = np.random.choice(all_doors)    # Define your primary choice
    
    print(f'Reward: {reward}\nPC: {primary_choice}')
    
    if primary_choice == reward:
        first_outcome.append(1)
        second_outcome.append(0)
        change.append(0)
        
        print(f'First step reward!')
        
    else:
        first_outcome.append(0)
        
        ## Monty opens the door
        monty_opens = all_doors.copy()
        monty_opens.remove(reward)
        monty_opens.remove(primary_choice)
        monty_opens = monty_opens[0]
        
        print(f'Monty opens: {monty_opens}')
        
        to_choose = all_doors.copy()
        to_choose.remove(monty_opens)
        
        secondary_choice = np.random.choice(to_choose)
        
        print(f'To choose: {to_choose}')
        print(f'SC: {secondary_choice}')
        
        if secondary_choice == primary_choice:
            change.append(0)
            print('Chnage: 0')
        else: 
            change.append(1)
            print('Change: 1')
            
        if secondary_choice == reward:
            second_outcome.append(1)
            print(f'Second step reward!')
        else: 
            second_outcome.append(0)
            print(f'No second step reward :(')
            
        
    print('\n')
```

```python
data = pd.DataFrame({'first_outcome': first_outcome, 
                     'second_outcome': second_outcome,
                     'change': change})
```
